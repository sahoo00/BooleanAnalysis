```python
def getChernov2025WBII(self, tn=1, ta=0):
    self.prepareDataDf("BL66.2")
    atype = self.getSurvName('c time')
    atypes = ['B', 'T1', 'T2', 'T3']
    ahash = {'Day0':0, 'Day3':1, 'Day5':2, 'Day7':3}
    if tn == 2:
        atypes = ['B', 'T1']
        ahash = {'Day0':0, 'Day3':1}
    if tn == 3:
        atypes = ['B', 'T2']
        ahash = {'Day0':0, 'Day5':1}
    if tn == 4:
        atypes = ['B', 'T3']
        ahash = {'Day0':0, 'Day7':1}
    self.initData(atype, atypes, ahash)
    return
BooleanAnalysis.getChernov2025WBII = getChernov2025WBII
```
